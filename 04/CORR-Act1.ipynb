c) ***Pour aller plus loin*** : écrire une méthode **`__str__`** pour la classe `Maillon`.

```
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def aff(self):
        print (self.valeur)
        if self.suivant != None:
            self.suivant.aff()
    
    def txt(self):
        if self.suivant is None : 
            return str(self.valeur) + " . "
        return str(self.valeur) + " , " + self.suivant.txt()
    
    def __str__(self):
        return self.txt()
        
    

m1 = Maillon(1)
m2 = Maillon(5,m1)
m3 = Maillon(12,m2)
```
