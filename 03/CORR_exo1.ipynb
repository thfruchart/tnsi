```python
class Eleve:
    def __init__(self, n,p,a):
        self.nom = n
        self.prenom = p
        self.age = a
    def affiche(self):
        print(self.prenom, self.nom, self.age)
    def __str__(self):
        return self.prenom + ' ' +  self.nom + ' ' + str(self.age) + ' ans'
        
bill = Eleve('GATES', 'Bill', 64)
moi = Eleve('MECHANT', 'Moi', 99)
```
