# Exercice 4
Écrire une fonction `tri_par_insertion(lst)` qui prend en argument une liste chaînée lst, et qui
renvoie une nouvelle liste chaînée contenant toutes les valeurs de lst triées dans l’ordre croissant
On suggère d’écrire cette fonction de manière récursive, en faisant appel à l’exercice précédent !

```python
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv
    
    def __repr__(self):
        if self is None:
            return ''
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)

def copie(lst):
    if lst is None : 
        return None
    # sinon lst est un Maillon
    return Maillon(lst.valeur, copie(lst.suivant))

def insere_trie(x, lst):
    if lst is None : 
        return Maillon(x)
    # sinon lst est un Maillon
    if x > lst.valeur :
        return Maillon(lst.valeur, insere_trie(x,lst.suivant))
    # sinon x <= lst.valeur et il faut insérer 2 Maillons
    return Maillon(x,  Maillon(lst.valeur, copie(lst.suivant)) )


lst = Maillon(30,Maillon(20, Maillon(50, Maillon(40))))
print(lst)
```

```python
def tri_par_insertion(lst) :
    if lst.suivant is None : # il n'y a qu'un seul Maillon
        return Maillon(lst.valeur) # copie de l'unique Maillon
    # else : 
    valeur_de_tete = lst.valeur
    trier_la_suite_de_liste = tri_par_insertion(lst.suivant) 
    return insere_trie(valeur_de_tete, trier_la_suite_de_liste) 
```

```python
tri_par_insertion(lst)
```

```python
def tri_par_insertion_iteratif(lst) :
    resu =  Maillon(lst.valeur) # copie du Maillon de têtê
    while lst.suivant is not None: # arrêt pour le dernier Maillon
        # on passe au Maillon suivant
        lst = lst.suivant
        resu = insere_trie(lst.valeur, resu)
    return resu 
```

```python
tri_par_insertion_iteratif(lst)
```

```python
def identiques(l1, l2) : 
    print(f"identiques({l1}, {l2})")
    if l1 is None and l2 is None : 
        return True
    if (l1 is None and l2 is not None): 
        return False
    if (l1 is not None and l2 is None): 
        return False
    if l1.valeur == l2.valeur : 
        return identiques(l1.suivant, l2.suivant)
    else : 
        return False
```

```python
identiques( Maillon(10,Maillon(20)) , Maillon(10,Maillon(20)) )
```

```python
identiques( Maillon(10,Maillon(20)) , Maillon(10,Maillon(30)) )
```
