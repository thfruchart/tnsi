## Exercice 2

Écrire une fonction `longueur(lst)` qui utilise une boucle **while** pour renvoyer la longueur d'une liste, c'est à dire le nombre de maillons qu'elle contient. On convient que `None` représente une liste de longueur 0. 

```
class Maillon:  # identique à l'activité 1
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

def mystere(lst):
    if lst is None:
        return 0
    return 1 + mystere(lst.suivant)

def longueur(lst):
    cpt  = 0 
    while lst is not None:
        cpt += 1
        lst = lst.suivant
    return cpt
```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
print (longueur (lst) )
```

### Exercice 2bis

En vous inspirant de l'exercice 2, écrire une fonction `total(lst)` qui renvoie le total des valeurs de tous les maillons d'une chaîne.

Solution récursive

```
class Maillon:  # identique à l'activité 1
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

def mystere(lst):
    if lst is None:
        return 0
    return 1 + mystere(lst.suivant)

def total(lst):
    if lst is None:
        return 0
    return lst.valeur + total(lst.suivant)

```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
print (total (lst) )
```

Solution itérative (while)

```
class Maillon:  # identique à l'activité 1
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

def longueur(lst):
    cpt  = 0 
    while lst is not None:
        cpt += 1
        lst = lst.suivant
    return cpt

def total(lst):
    cpt  = 0 
    while lst is not None:
        cpt += lst.valeur
        lst = lst.suivant
    return cpt
```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
print (total (lst) )
```

# Transformer les fonctions en méthodes

```
class Maillon:  # identique à l'activité 1
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def longueur(self):
        cpt  = 0 
        while self is not None:
            cpt += 1
            self = self.suivant
        return cpt

    def total(self):
        cpt  = 0 
        while self is not None:
            cpt += self.valeur
            self = self.suivant
        return cpt
```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
print (lst.longueur() )
```

```
lst.total()
```

2) Solution avec fonctions récursives

```
class Maillon:  # identique à l'activité 1
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def longueur(self):
        if self.suivant is None:
            return 1
        return 1 + (self.suivant).longueur()

    def total(self):
        if self.suivant is None:
            return self.valeur
        return self.valeur + (self.suivant).total()

```

```
lst = Maillon(30, Maillon(20, Maillon(10)))
print (lst.longueur() )
```

```
lst.total()
```
