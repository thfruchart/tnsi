```python
class Carte : 
    def __init__(self, v : int) : 
        self.valeur = v
        self.TdB = self.calcul_TdB()
    
    def calcul_TdB(self) : 
        resu = 0
        v = self.valeur
        if v % 11 == 0 : # v est multiple de 11
            resu += 5
        if v % 10 == 0 : # finit par 0
            resu += 3
        elif v % 5 == 0 : # finit par 5
            resu += 2
        if resu == 0 :
            resu = 1
        return resu 
    
    def est_superieure_a(self , autre) : 
        v1 = self.valeur
        v2 = autre.valeur
        return v1>v2
    
    # BONUS
    def difference(self , autre) : 
        v1 = self.valeur
        v2 = autre.valeur
        return abs(v1-v2)

class Paquet :
    def __init__(self,L):
        self.contenu = L
        
    def afficher(self) : 
        for carte in self.contenu : 
            print(carte.valeur)
        
    def ajouter_carte(self, carte) :
        self.contenu.append(carte)
        
    def nombre_TdB(self) : 
        total = 0
        for cart in self.contenu : # cart est de type Carte
            total += cart.TdB
        return total
    
    def distribuer(self, nbr) :
        assert nbr <= 10
        liste_de_paquets = [ ]
        for i in range(nbr) : 
            liste_de_paquets.append( Paquet([]) )
        # liste_de_paquets = [ Paquet([]) for i in range(nbr) ]
        for _ in range(10) :# on répète 10 fois l'opération
            #il faut ajouter 1 carte à chaque paquet de la liste_de_paquets
            for paq in liste_de_paquets: 
                new_card = self.contenu.pop()
                paq.ajouter_carte(new_card)
        return liste_de_paquets
        
        
        
    
        
        
```

```python
monpaquet = Paquet([Carte(11), Carte(50)])
```
