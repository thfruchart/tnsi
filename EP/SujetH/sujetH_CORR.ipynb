```python
a = {'F':['B','G'], 'B':['A','D'], 'A':['',''], 'D':['C','E'], 'C':['',''], 'E':['',''], 'G':['','I'], 'I':['','H'], 'H':['','']}
```

```python
def taille(arbre,lettre):
    if arbre[lettre] == ['','']:
        return 1
    elif arbre[lettre][0] == '' :
        return 1 + taille(arbre, arbre[lettre][1] )
    elif arbre[lettre][1] == '' :
        return 1 + taille(arbre, arbre[lettre][0] )
    else : 
        return 1 + taille(arbre, arbre[lettre][0] ) +  taille(arbre, arbre[lettre][1] )
```

```python
taille(a,'F')
```

```python
taille(a,'B')
```

```python
taille(a,'E')
```

```python
taille(a,'F')
```

```python
def tri_iteratif(tab):
    for k in range( len(tab)-1 , 0, -1): 
        imax = 0
        for i in range(0 , k ):
            if tab[i] > tab[imax] :
                imax = i
        if tab[imax] > tab[k] :
            tab[k] , tab[imax] = tab[imax] , tab[k]
    return tab
```

```python
tri_iteratif([40,20,10,50,70])
```

```python
tri_iteratif([100,40,20,10,50,70])
```
