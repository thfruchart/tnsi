Ecrire une fonction `profondeur(arbre)` qui prend en argument un objet de type `Noeud` et qui renvoie la profondeur de `arbre`.

Exemple  :  

    >>> profondeur(Noeud( Noeud(None,'A',None), 'racine', Noeud(None,'B',None) ))
    2
    >>> profondeur(Noeud(None,'A',None) )
    1


```
class Noeud:
    def __init__(self, g, v, d):
        self.gauche = g
        self.valeur = v        
        self.droit  = d

arbre1 = Noeud( Noeud(None,'A',None), 'racine', Noeud(None,'B',None) )
arbre1.gauche.droit = Noeud(None,'C',None)
arbre2 = Noeud( Noeud(None,'A',None), 'racine', Noeud(None,'B',None) )
arbre2.droit.gauche = Noeud(None,'C', Noeud(None,'D',None))


def profondeur(arbre):
    if arbre is None : 
        return 0
    cpt = 1 #racine 
    cpt_gauche = profondeur(arbre.gauche)
    cpt_droit = profondeur(arbre.droit)
    print(f"cpt_gauche = {cpt_gauche}, cpt_droit={cpt_droit}")
    return 1 + max (cpt_gauche, cpt_droit)

# ou plus simplement 
def profondeur(arbre):
    if arbre is None : 
        return 0
    return 1 + max (profondeur(arbre.gauche), profondeur(arbre.droit))
```
