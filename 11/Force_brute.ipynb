# Fonction de hachage

```python
import hashlib

def hache(txt :str):
    m = hashlib.sha256(txt.encode())
    return m.hexdigest()
```

## exemple de texte "haché"

```python
hache('pass')
```

```python
hache('passe')
```

```python
hache('Pass')
```

```python
hache('bzm05')
```

```python
hache('tp88bts')
```
