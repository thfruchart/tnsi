question 2) 
Pour ajouter un nouveau maillon à la fin de la chaîne

```python
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv
lst = Maillon(10,Maillon(20))
################################
lst.suivant = Maillon( 15 , lst.suivant)
m = lst
while m.suivant is not None :   # arrêt lorsque m.suivant est None
    m = m.suivant
m.suivant = Maillon(25)
```
