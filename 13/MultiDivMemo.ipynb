## utilisation de la programmation dynamique
#### premier essai

```python
N = 16
terrain = [i for i in range(1,N+1)]
nb_rec = 0

def accessible(a,b):
    return a%b==0 or b%a==0

memo = {}

def plus_long_chemin_memo(chemin_courant):
    '''chemin_courant est un TUPLE contenant un chemin valide
    la fonction renvoie un plus long tuple obtenu en ajoutant de nouvelles valeurs'''
    global nb_rec
    # cas de base
    if chemin_courant in memo:
        return memo[chemin_courant]
    # cas récursif
    chemin_optimal = chemin_courant
    nombres_restant = [i for i in terrain if i not in chemin_courant]
    
    for suiv in nombres_restant:
        if len(chemin_courant)==0 or accessible(chemin_courant[-1],suiv):
            nb_rec+=1
            chemin = plus_long_chemin_memo(chemin_courant+(suiv,))
            if len(chemin) > len(chemin_optimal):
                chemin_optimal = chemin
    memo[chemin_courant] = chemin_optimal
    
    return chemin_optimal  
```

### amélioration des paramètres : 
valeur de départ et liste de nombres restant

```python
N = 12
terrain = [i for i in range(1,N+1)]

def accessible(a,b):
    return a%b==0 or b%a==0

rec = 0  # compte les appels récursifs

def plus_long_chemin_depuis(depart, valeurs_restantes):
    '''valeurs_restantes est un tuple de valeurs disponibles
    la fonction renvoie une plus longue liste qui commence avec depart
    et se poursuit avec des valeurs restantes si possible'''
    global rec
    #print(f"plus_long_chemin_depuis({depart}, {valeurs_restantes})")
    chemin_optimal = [depart]
    for suiv in valeurs_restantes:
        if accessible(depart,suiv):
            rec+=1
            chemin = [depart] + plus_long_chemin_depuis(suiv, tuple([v for v in valeurs_restantes if v != suiv]))
            if len(chemin) > len(chemin_optimal):
                chemin_optimal = chemin
    
    return chemin_optimal 

def plus_long_chemin_memo(terrain):
    global rec
    cpt=0
    rec=0
    optimal = []
    for v in terrain:
        chemin = plus_long_chemin_depuis(v, tuple([k for k in terrain if k != v]))
        if len(chemin) > len(optimal):
            optimal = chemin
    return optimal
        
```

# Ajouter la "mémoïsation" dans un dictionnaire 
### objectif : diminuer le nombre d'appels récursifs

```python
N = 12
terrain = [i for i in range(1,N+1)]

def accessible(a,b):
    return a%b==0 or b%a==0

rec = 0  # compte les appels récursifs
cpt = 0  # compte les accès au dictionnaire memo

memo = {}

def plus_long_chemin_depuis(depart, valeurs_restantes):
    '''valeurs_restantes est un tuple de valeurs disponibles
    la fonction renvoie une plus longue liste qui commence avec depart
    et se poursuit avec des valeurs restantes si possible'''
    global cpt, rec
    if (depart, valeurs_restantes) in memo:
        cpt += 1
        return memo[(depart, valeurs_restantes)]
    chemin_optimal = [depart]
    for suiv in valeurs_restantes:
        if accessible(depart,suiv):
            rec+=1
            chemin = [depart] + plus_long_chemin_depuis(suiv, tuple([v for v in valeurs_restantes if v != suiv]))
            if len(chemin) > len(chemin_optimal):
                chemin_optimal = chemin
    memo[(depart, valeurs_restantes)] = chemin_optimal
    return chemin_optimal 

def plus_long_chemin_memo(terrain):
    global cpt, rec
    cpt=0
    rec=0
    optimal = []
    for v in terrain:
        chemin = plus_long_chemin_depuis(v, tuple([k for k in terrain if k != v]))
        if len(chemin) > len(optimal):
            optimal = chemin
    return optimal
        
```
