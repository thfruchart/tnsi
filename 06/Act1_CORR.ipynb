|fonction  |description|

|`sommet(p:Pile[T]) -> T`|renvoie l'élément au sommet de `p` sans le retirer|
|`depiler(p:Pile[T]) -> T` | retire et renvoie l'élément au sommet de `p`|

#### a) programmer la fonction sommet

```
def sommet(p):
    maillon_du_haut = p.haut
    return maillon_du_haut.valeur # p.haut.valeur
```

#### b) programmer la fonction depiler

```
def depiler(p):
    assert not p.est_vide(), "une pile vide ne peut pas être dépilée"
    resu = sommet(p) # p.haut.valeur
    p.haut = p.haut.suivant
    return resu  
```

# classe Pile complète

```
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)

class Pile:
    def __init__(self):
        self.haut = None

    def est_vide(self):
        return self.haut is None

    def empiler(self, val):
        self.haut = Maillon(val, self.haut)

    def __str__(self):
        return "HAUT:"+str(self.haut)+'.'

    def depiler(self):
        assert not self.est_vide(), "une pile vide ne peut pas être dépilée"
        resu = self.haut.valeur
        self.haut = self.haut.suivant
        return resu 
```

```
p = Pile()
p.empiler(5)
p.empiler(10)
p.empiler(20)

assert sommet(p)==20
assert p.depiler() == 20
assert p.depiler() == 10 
assert p.depiler() == 5
assert p.est_vide()
print("test ok")
p.depiler()
```
