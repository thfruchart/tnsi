## Exercice 1

Ecrire une fonction `identiques(a1, a2)` qui permet de tester si les deux objets `a1` et `a2`de type `Noeud`définissent des arbres identiques. 

```
class Noeud:
    def __init__(self, g, v, d):
        self.gauche = g
        self.valeur = v        
        self.droit  = d

def identiques(a1, a2):
    """a1 et a2 sont de type Noeud
    renvoie True si a1 et a2 définissent des arbres binaires identiques
    renvoie False sinon"""
    if a1 is None and a2 is None : 
        return True
    # else : a1 et a2 ne sont pas TOUS DEUX None
    if a1 is None or a2 is None : 
        return False
    # aucun des deux n'est None
    if a1.valeur != a2.valeur:
        return False
    return identiques(a1.gauche, a2.gauche) and identiques(a1.droit, a2.droit)
    
```

```
a1 = Noeud(None,'R', None)
a2 = Noeud(Noeud(None,'G', None),'R', Noeud(None,'D', None))
a3 = Noeud(Noeud(Noeud(None,'gg', None),'G', Noeud(None,'gd', None)),
           'R', 
           Noeud(Noeud(None,'dg', None),'D', Noeud(None,'dd', None)))
a2bis = Noeud(Noeud(None,'G', None),'R', Noeud(None,'D', None))
```

```
identiques(a1,a2)
```
