```python
arb = ["A", [["B", [["E", []]]], ["C", []], ["D", []]]]
```

```python
def taille(arbre):
    if len(arbre) == 0:
        return 0
    resu = 1 # pour la racine de l'arbre
    print ('racine : ', arbre[0])
    enfants = arbre[1]
    for sous_arbre in enfants :
        resu += taille(sous_arbre)
    return resu
```

```python
taille(arb)
```
