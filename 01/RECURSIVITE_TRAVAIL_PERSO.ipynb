## Calcul "à la main"

On donne le code qui définit deux fonctions récursives :  
`compteur1`  et `compteur2`

```
def compteur1(n):
    if n>=0:
        print(n)
        compteur1(n-1)
```

```
def compteur2(n):
    if n>=0:
        compteur2(n-1)
        print(n)
```
