```python
class Carte : 
    def __init__(self, v : int) : 
        self.valeur = v
        self.TdB = self.calcul_TdB()
    
    def calcul_TdB(self) : 
        resu = 0
        v = self.valeur
        if v % 11 == 0 : # v est multiple de 11
            resu += 5
        if v % 10 == 0 : # finit par 0
            resu += 3
        elif v % 5 == 0 : # finit par 5
            resu += 2
        if resu == 0 :
            resu = 1
        return resu 
    
    def est_superieure_a(self , autre) : 
        v1 = self.valeur
        v2 = autre.valeur
        return v1>v2
    
    # BONUS
    def difference(self , autre) : 
        v1 = self.valeur
        v2 = autre.valeur
        return abs(v1-v2)

class Paquet :
    def __init__(self,L):
        self.contenu = L
        
    def afficher(self) : 
        for carte in self.contenu : 
            print(carte.valeur)
        
    def ajouter_carte(self, carte) :
        self.contenu.append(carte)
        
        
    
        
        
```

```python
monpaquet = Paquet([Carte(11), Carte(50)])
```
