# Activité 3

On reprend la classe Maillon de l'activité 1, et on définit une fonction `mystere`

```
class Maillon:  
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

def mystere(lst, n):
    if n == 0:
        return lst.valeur
    else:
        return mystere(lst.suivant, n-1)
```

### Exemple 1

la valeur affiché est celle du maillon de tête (car n vaut 0)

```
lst = Maillon(30, Maillon(20, Maillon(10)))
x = mystere(lst, 0)
print(x)
```

```
lst = Maillon(50, Maillon(60, Maillon(70)))
y = mystere(lst, 2)
print(y)
```

### Exemple 3

```
lst = Maillon(1, Maillon(2, Maillon(3)))
z = mystere(lst, 3)
print(z)
```

La fonction mystère renvoie, si elle existe, la valeur du maillon n° `n`.
Le maillon de tête correspond à la valeur `n = 0`
Si n est trop grand, une `ValueError` est levée.

```
class Maillon:  
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

def mystere(lst, n):
    if lst is None : 
        raise ValueError("n doit être inférieur au nombre de maillons de la chaîne")
    if n == 0:
        return lst.valeur
    else:
        return mystere(lst.suivant, n-1)
```

```
lst = Maillon(1, Maillon(2, Maillon(3)))
z = mystere(lst, 3)
```
