# Exercice 2
Écrire une fonction occurrences(x, lst) qui renvoie le nombre d’occurrences d’une valeur x dans
une liste chaînée lst.

a) avec une fonction récursive

b) avec une boucle while

```python
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv
    
    def __str__(self):
        if self is None:
            return ''
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)

lst = Maillon(10,Maillon(20, Maillon(10, Maillon(30))))
print(lst)
```

question a)

```python
def occurrences(x, lst) : # version récursive
    if lst.suivant is None : 
        if x == lst.valeur : 
            return 1
        else : 
            return 0
    else : 
        if x == lst.valeur : 
            return 1 + occurrences(x, lst.suivant)
        else : 
            return 0 + occurrences(x, lst.suivant)
```

```python
def occurrences(x, lst) : # version récursive améliorée
    if lst is None : # pas de maillon
        return 0
    # cas récursif
    if x == lst.valeur : 
        cpt =  1 
    else : 
        cpt =  0 
    return cpt + occurrences(x, lst.suivant)
```

```python
occurrences(10,lst), occurrences(30,lst), occurrences(40,lst)
```

question b)

```python
def occurrences(x, lst) : # version itérative
    cpt =  0 
    while lst is not None : 
        if x == lst.valeur : 
            cpt +=  1 
        lst = lst.suivant
    return cpt
            
```

```python
occurrences(10,lst), occurrences(30,lst), occurrences(40,lst)
```
