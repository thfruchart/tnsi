```python
def rechercheMinMax(tab):
    d = {}
    d['min'] = None
    d['max'] = None
    if tab == [] : 
        return d
    # tableau non vide
    d['min'] = tab[0]
    d['max'] = tab[0]
    for v in tab :
        if d['max'] < v:
            d['max'] = v
        if d['min'] > v:
            d['min'] = v
    return d
 
```

```python
def rechercheMinMax(tab):
    d = {}
    d['min'] = None
    d['max'] = None
    if tab == [] : 
        return d
    # tableau non vide
    for v in tab :
        if d['max'] is None or d['max'] < v:
            d['max'] = v
        if d['min'] is None or d['min'] > v:
            d['min'] = v
    return d
 
```

```python
class Carte:
    """Initialise Couleur (entre 1 à 4), et Valeur (entre 1 à 13)"""
    def __init__(self, c, v):
        self.Couleur = c
        self.Valeur = v

    """Renvoie le nom de la Carte As, 2, ... 10, 
       Valet, Dame, Roi"""
    def getNom(self):
        if ( self.Valeur > 1 and self.Valeur < 11):
            return str( self.Valeur)
        elif self.Valeur == 11:
            return "Valet"
        elif self.Valeur == 12:
            return "Dame"
        elif self.Valeur == 13:
            return "Roi"
        else:
            return "As"

    """Renvoie la couleur de la Carte (parmi pique, coeur, carreau, trefle"""
    def getCouleur(self):
        return ['pique', 'coeur', 'carreau', 'trefle' ][self.Couleur - 1]
    
    def __repr__(self):
        return self.getNom() + " de " + self.getCouleur()
    

class PaquetDeCarte:
    def __init__(self):
        self.contenu = []

    """Remplit le paquet de cartes"""
    def remplir(self):
        
        for couleur in range(1,5) :
            for valeur in range(1,14) :
                nouvelle_carte = Carte(couleur, valeur)
                #print(nouvelle_carte.getNom() + " de " + nouvelle_carte.getCouleur())
                self.contenu.append(nouvelle_carte)
        return 

    

    """Renvoie la Carte qui se trouve à la position donnée"""
    def getCarteAt(self, pos):
        return self.contenu[pos]
        
```

```python
p = PaquetDeCarte()
```

```python
p.remplir()
```
